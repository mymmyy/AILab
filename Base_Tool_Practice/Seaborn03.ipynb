# 回归分析（连续值）

```python
%matplotlib inline
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt

import seaborn as sns
sns.set(color_codes=True)
np.random.seed(sum(map(ord,"regression")))

# 默认数据集
tips = sns.load_dataset("tips")
tips.head()
```

## regplot() 和 implot()都可以绘制回归关系，推荐regplot()

```python
sns.regplot(x='total_bill', y='tip', data=tips)
```

```python

# x_jitter 浮动范围
sns.regplot(x='size', y='tip', data=tips, x_jitter=.05)
```

# 类别数据集回归

```python
sns.set(style="whitegrid", color_codes=True)
np.random.seed(sum(map(ord, "categorical")))
titanic = sns.load_dataset("titanic")
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
```

```python
# 类别
sns.stripplot(x="day", y='total_bill', data=tips)
```

```python
sns.stripplot(x="day", y='total_bill', data=tips, jitter=True)
```

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips)
```

## 盒图
- IQR即统计学概念四分位距，第一四分位与第三四分位之间的距离
- N = 1.5IQR 如果一个值> Q3+N或 < Q1-N 则为离群点

```python
sns.boxplot(x="day", y='total_bill', hue='time', data=tips)
```

```python
# 小提琴图
sns.violinplot(x='total_bill', y='day', hue='time', data=tips)
```

```python
# 小提琴图
sns.violinplot(x='total_bill', y='day', hue='time', data=tips,split=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, inner=None)
sns.swarmplot(x='day', y='total_bill', data=tips, color='w', alpha=0.5)
```

```python
# 显示值的集中趋势可以用条形图
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

```python
# 点图可以更好描述变化差异
sns.pointplot(x='sex', y='survived', hue='class', data=titanic)
```

# 分类属性

```python
sns.pointplot(x='class', y='survived', hue='sex', data=titanic, palette={'male':'g', 'female':'m'}, markers=['^', 'o'], linestyles=['-', '--'])
```

```python
# 宽型数据
sns.boxplot(data=iris, orient='h')
```

```python
# 多层面板分类图
sns.factorplot(x='day', y='total_bill', hue='smoker', data=tips)
```

```python
sns.factorplot(x='day', y='total_bill', hue='smoker', data=tips, kind='bar')
```

```python
sns.factorplot(x='day', y='total_bill', hue='smoker', data=tips, kind='swarm', col='time')
```

```python
sns.factorplot(x='time', y='total_bill', hue='smoker', data=tips, kind='box', col='day', size=4, aspect=.5)
```
